Suction caissons are installed by evacuating water from under a steel, tubular foundation with a sealed top (@fig-installation-process). Pumping lowers the water pressure inside the unit, creating a pressure differential that pushes the unit into the seabed. Seepage flow may also change the effective stresses inside and outside the caisson, depending on the permeability of the soil. 

![Installation procedure for a suction caisson](pics/suction/suction-installation.png){#fig-installation-process}

As mentioned in the [introductory notes](https://triaduct.github.io/offshoregeotechnics/notes/1_intro.html), foundations can be loosely categorised based on their slenderness ($L/D$), such as bucket foundations, caissons or piles. For consistency, this chapter will solely refer to the foundations as "suction caissons".

::: {.callout-note title="Note on standardisation"}
Suction-assisted installation is not covered in the ISO standards. As an alternative, there are two guidance documents that are often used: 

* Carbon Trust Offshore Wind Accelerator: Suction Installed Caisson Foundations for Offshore Wind: Design Guidelines 2019 [(link)](www.carbontrust.com/sites/default/files/documents/resource/public/owa-suction-caisson-design-guidelines-report.pdf) 
* DNV-RP-E303: Geotechnical design and installation of suction anchors in clay [(link)](https://www.dnv.com/energy/standards-guidelines/dnv-rp-e303-geotechnical-design-and-installation-of-suction-anchors-in-clay/)
:::

## Types of suction foundations
Suction caissons can be integrated within the superstructure itself (@fig-borkum-riffgrund). This greatly simplifies the construction process, since less lifts are needed compared to an equivalent piled jacket. Jacket inclination can also be controlled during installation by adjusting the suction pressures at each leg. 

::: {#fig-integrated-suction-caissons layout-ncol=2}
![Installation of jacket foundations at Borkum Riffgrund [(source)](https://www.sptoffshore.com/wind-turbine-generator-wtg-foundations/)](pics/suction/borkum-riffgrund.jpg){#fig-borkum-riffgrund}

![Monobucket foundations [(source)](https://www.windpowermonthly.com/article/1361921/gallery-installing-removing-universal-foundations-mono-buckets)](pics/suction/monobucket.jpg){#fig-monobuckets}

Examples of suction caissons integrated with the superstructure itself
:::

Suction caissons are also often used for temporary infrastructure, like pile guide frames (@fig-pile-guide-frame), mooring anchors or seabed infrastructure for oil and gas infrastructure. Not only can the caissons be easily installed into the shallow seabed, they can also be extracted easily by applying a backpressure (i.e. the opposite of suction pressure). This backpressure lifts the caisson out of the ground, allowing it to be reused at the next location.

![Lifting of a pile guide frame for Taiwanese project](pics/suction/pile-guide-frame.jpg){#fig-pile-guide-frame}



## Self-weight penetration
Predicting self-weight penetration is especially important for suction-assisted installation as it also tells us what kind of pump capacity we need, as well as anticipating the potential pore pressure field once suction is applied (particularly with respect to the influence of layering).

At a basic level, self-weight penetration can be accounted for by equating the downward forces (such as effective weight of the caisson) to the static resistance to driving (@fig-theoretical-example). This is somewhat of a simplification since the extent of the penetration is dependent on the caisson's speed and momentum---but nevertheless provides a decent first estimate. 

Just like pile driveability studies, a best estimate and high estimate prediction is made to ensure sufficient pump capacity is available.

![Theoretical example of suction caisson installation](pics/suction/theoretical-example.png){#fig-theoretical-example}



## Clay
To consider how a suction pressure $s$ affects the penetration of the caisson in @fig-caisson-suction-clay, let's look at the balance between upward and downward forces. We can do this using the total stress method for piles:

$$
V' = \underbrace{h\alpha_0s_{u1}(\pi D_o)}_{\text{Outer shaft friction}} + \underbrace{h \alpha_i s_{ui} (\pi D_i)}_{\text{Inner shaft friction}} + \underbrace{(\gamma' h N_q + s_{u2} N_c)(\pi D t)}_{\text{End-bearing on annulus}}
$$ {#eq-total-stress-method}

where $s_{u1}$ is the average undrained shear strength across the depth $h$, $s_{u2}$ is the undrained shear strength at the bottom of the caisson, $N_q$ is an overburden factor (=1 for undrained conditions), $N_c$ is a cohesion capacity factor (≈9) and $\alpha$ is an adhesion factor (=$1/S_t$ where $S_t$ is the soil sensitivity. 


![Caisson schematic](pics/suction/caisson-suction-clay.png){#fig-caisson-suction-clay width=70%}

Between the top of the caisson and the mudline, a hydrostatic pressure is acting on all sides of the caisson. By sucking the water out from between the caisson and the mudline, a pressure differential is created between the outside and inside of the caisson, creating a net downward force acting in the same direction as the caisson's effective weight:

$$
V' \textcolor{red}{+ s(\frac{\pi D_i^2}{4})} = h \alpha_0 s_{u1} (\pi D_o) + h \alpha_i s_{ui} (\pi D_i) + (\gamma' h N_q \textcolor{red}{-s} + s_{u2} N_c)(\pi D t)
$$ {#eq-suction-clay}

Since the lower pressure also reduces the overburden pressure, the surcharge term is also reduced. Most importantly, it is assumed that no seepage flow occurs in clay given its high impermeability.


## Sand
Suction installation in sand creates a seepage flow from outside the caisson, around the caisson tip and up the inside of the caisson (@fig-caisson-suction-sand). Therefore, the conditions are a little different compared to clay.

The seepage flow changes the effectives stress in the soil. Seepage flow leads to a decrease in the effective stress inside the caisson and an increase in the effective stress outside the caisson. However the most important mechanisms is how it reduces the resistance acting on the caisson tip because of the lower overburden stress contribution provided by soil plug (@fig-caisson-suction-sand).


:::{.callout-tip title="Why do the effective stress change inside and outside the caisson?" collapse="true"}
### Downward flow
Consider the stresses in @fig-downward-flow, where water flowing into the top of the tank and falling by gravity through Layer 1 and Layer 2.

![Downward flow](pics/suction/downward-flow.png){#fig-downward-flow width=60%}

|  | Point A (Beginning) | Point B (Middle) | Point C (End)  |
|----|-----|----|----|
| Total stress $\sigma$       | $\gamma_w H_w$           | $\gamma_w H_w + \gamma_{sat,1} H_1$ | $\gamma_w H_w + \gamma_{sat,2} H_1 + \gamma_{sat,1} H_2$  |
| Pore pressure $u$           | $\gamma_w H_w$           | $\gamma_w H_{w1}$                   | 0  |
| Effective stress $\sigma'$  | 0                        | $\sigma'_B > \sigma'_A$             | $\color{red}{\sigma'_C > \sigma'_B > \sigma'_A}$  |

The end-result at Point C implies that **we get an increase in effective stress along downward flow paths**.

- - - 

### Upward flow
Now consider the opposite scenario, where the setup in @fig-upward-flow allows us to consider the stress along an upward flow path.

![Upward flow](pics/suction/upward-flow.png){#fig-upward-flow width=60%}

|  | Point C (Beginning) | Point B (Middle) | Point A (End)  |
|----|-----|----|----|
| Total stress $\sigma$       | $\gamma_w H_w + \gamma_{sat,1} H_1 + \gamma_{sat,2} H_2$ | $\gamma_w H_w + \gamma_{sat,1} H_1$ | $\gamma_w H_w$  |
| Pore pressure $u$           | $\gamma_w H_w2$                                          | $\gamma_w H_{w1}$                   | $\gamma_w H_w$  |
| Effective stress $\sigma'$  | $\color{red}{\sigma'_C > \sigma'_B > \sigma'_A}$}        | $\sigma'_B > \sigma'_A$             | $\sigma'_A = 0$ |

Now we have the opposite effect compared to the downward flow scenario: **there is a decrease in effective stress along the upward flow path**.

:::

Delving into the mechanisms of suction-assisted installation a bit further, any time we have flow through a medium, we also have a loss of energy. As stated by Darcy's Law:

$$
v = k \frac{\Delta h}{L} = ki
$$

where $k$ is the permeability, $\Delta h$ is the change in hydraulic head, $L$ is the length of the flow path and $i$ is the hydraulic gradient. 

Ideally, we would model the seepage path as a function of the permeability $k$, but there's a couple of problems in doing so when modelling a penetrating caisson:

1. The density of the soil changes both inside and outside the caisson---and therefore also the permeability
2. The penetrating caisson also affects the seepage path length 

As one solution to this problem, @houlsbyDesignProceduresInstallation2005 proposed taking the absolute pressure at the caisson tip as:

$$
p_{tip} = p_a + \gamma_w(h_w + h) - \textcolor{red}{a}s
$$

where $a$ is a factor that describes the reduction in suction pressure between the top and bottom of the caisson. This means that the upward hydraulic gradient $i$ inside the caisson is $(1-a)s/\gamma_w h$ and the downward hydraulic gradient is $as/\gamma_w h$.

If $a=0.50$, then the hydraulic gradient both inside and outside the caisson is the same---and likewise so is the change in effective stress. 

However, flow within the caisson is much more constrained compared to outside the caisson and so $a$ is often less than 0.50. This means that there's a higher hydraulic gradient inside the caisson then outside and so the **reduction in effective stress inside the caisson is greater than the increase in effective stress outside the caisson**.

:::{.callout-tip title="Quick note on the pore pressure factor $a$" collapse="true"}
Of course, it isn't always that simple. Consider the numerical studies shown in @fig-pore-pressure-factor, which shows that $a$ even changes with the caisson penetration. This also shows results as a function of $k_f$ which describes the initial permeabilities both inside and outside the caisson ($k_f = k_i/k_o$).

![Variation of the pore pressure factor $a$ with $h/d$ and $k_f$ [@houlsbyDesignProceduresInstallation2005]](pics/suction/houlsby-fig4.png){#fig-pore-pressure-factor width=60%}
:::

To consider this in an effective stress method for piles, let's first look at the basic equation for a pile (or self-weight penetration of a caisson): 

$$
V' = 
\underbrace{\frac{\gamma'h^2}{2} (K \tan{\delta})_o(\pi D_o)}_{\text{Outer shaft friction}} + 
\underbrace{\frac{\gamma'h^2}{2} (K \tan{\delta})_i(\pi D_i)}_{\text{Inner shaft friction}} + 
\underbrace{(\gamma' h N_q + \gamma' \frac{t}{2} N_\gamma )(\pi D t)}_{\text{End-bearing on annulus}}
$$ {#eq-effective-stress-method}

In this equation, the shaft friction is modelled as the total applied horizontal stress on the outer and inner walls (i.e. $\sigma'_v K \tan{\delta}$). To adjust this equation to account for seepage flow, we adjust the stress term $\sigma'_v$. For simplicity, we'll just adjust the unit weight term whereby:

$$
\gamma'_o = \gamma'  + \frac{as}{h}
$$

$$
\gamma'_i = \gamma'  - \frac{(1-a)s}{h}
$$

Just like clay, the left-hand side of the equation is also adjusted to accounted for the lower pressure between the mudline and the top of the caisson. The overburden stress will also need to adjusted in the end-bearing term, leading to:

$$
V' \textcolor{red}{+ s(\frac{\pi D_i^2}{4})} = 
\frac{\textcolor{red}{\gamma'_o}h^2}{2} (K \tan{\delta})_o(\pi D_o) + 
\frac{\textcolor{red}{\gamma'_i}h^2}{2} (K \tan{\delta})_i(\pi D_i) + 
(\textcolor{red}{\gamma'_i} h N_q + \gamma' \frac{t}{2} N_\gamma)(\pi D t)
$$ {#eq-effective-stress-method}

Keep in mind that this equation is very much a simplification of what's going on. Differences in effective stress both inside and outside the caisson will affect the asymmetry of the distributed load on the annulus. Moreover, the equation also does not account for the enhancement in stress caused by caisson displacing the soil laterally, as well as the internal restraint on the plug. 

More detail into analytical formulations for suction caissons is provided in @houlsbyDesignProceduresInstallation2005.

## Layered soils 
Of course, homogeneous sand or clay layers are seldom found in practice, especially for long suction piles that penetrate to depths of up to 30m. 

For a more detailed assessment, particularly regarding installation prediction, check out @klinkvortPenetrationModelInstallation2019.

### Clay overlying sand
A common assumption is that seepage flow won't develop when a clay layer overlies sand, since the clay acts as an impermeable barrier between the pump and underlying sand. However, there are a couple of reasons as to why seepage flow through a clay layer overlying sand, and this all depends on the clay's stiffness and thickness.

Firstly, a reduction in the compartment pressure may lead to uplift of the clay layer itself (@fig-uplift). This will happen when the pressure underneath the clay layer exceeds both the compartment pressure and the adhesion between the clay layer and the caisson. This is often accompanied by gapping between the clay layer and the skirt, allowing for a seepage path to develop around the clay layer. 

Alternatively, if the pressure differential on either side of the clay layer gets high enough, the clay layer might crack---especially for stiff, fissureable clay. This can also happen simultaneously with plug uplift.

![Plug uplift observed during particle image velocimetry (PIV) tests by @ragniObservationsEffectsClay2020](pics/suction/plug-uplift.png){#fig-uplift width=80%}

### Sand overlying clay
When sand overlies clay, no flow would expected once the caisson reaches the clay layer. Therefore, the effect of seepage flow on reducing the effective stress inside the caisson and underneath the caisson tip will not present. 

However, since the clay layer often provides a lower resistance compared to sand, some additional penetration through the clay layer would be expected---although this depends on the relative contributions of the soil resistance and pressure differential between inside and outside the caisson. 

## CPT-based design
CPT-based methods for forecasting suction caisson installation are comparable to SRD formulations for piles---and often the SRD methods are used themselves with some modifications.

DNV-RP-E303 provides a method designed specifically for suction caissons, based on a database of stiff clays and dense sands in the North Sea: 

$$
R_b(z) = A_{tip} k_p(z) q_c(z) 
$$

$$
R_s(z) = A_{wall}(z) \int_0^z k_f(z) q_c(z) dz
$$

## Limitations to suction installation
### Cavitation
**Cavitation** describes the absolute pressure at which water turns into a vapour, resulting in the prdocution of cavities (or bubbles) in the water body. When these cavities go from a region of low pressure to high pressure through the pump, they may suddenly collapse---generating shockwaves which can severely damage the pump itself. 

For sea temperatures of 10-20°C, the cavitation point would be expected at around 1 to 2 kPa. Therefore, for installation we need to ensure that:

$$
p_a + \gamma_w h_w - s \ge 2 \text{kPa}
$$ {#eq-cavitation}

### Piping
Piping (also known as ratholing) is one of the most critical limitations to suction installation in sand. Piping occurs when the upward seepage pressure equals the effective unit weight:

$$
\gamma' = \frac{\gamma_w (1-a)s}{\gamma_w h}
$$ {#eq-piping}

Once this point is reached, a channel of sand particles is lifted up and taken along with the flow path, resutling in a complete destructuration of soil both inside and outside the caisson. 

In some sense, piping can be viewed as beneficial for installation, since the effective stress inside and outside the caisson is significantly reduced. However, piping can create excessive inclination in the foundation itself and it also becomes difficult to apply suction again once piping occurs. Moreover, there'll be even more uncertainty regarding the soil conditions around the caisson, potentially creating issues for the capacity of the structure after installation.

All-in-all, suction-assisted installation is about striking a balance between creating enough seepage, without letting piping occur

![Piping underneath a model dam [(Real Engineering)](https://www.youtube.com/watch?v=eImtYyuQCZ8)](pics/suction/piping.gif){#fig-piping width=70%}

### Structural failure
**Structural failure** may occur if the water and lateral earth pressure outside the caisson exceeds that within the caisson. Webs and stiffners are often included on the caisson to prevent this, although these may also add an additional resistance component, particularly in the case of joined stiffners (@fig-webs-stiffeners).
